```python
with open('64317-0.txt', mode = 'r', encoding = 'utf-8') as f:
    lines = f.readlines()
```

```python
len(lines)
```

```python
lines[0]
```

```python
Novel = []
for i in range(0,6773):
    a = lines[i]
    a = a.replace('\n', '')
    a = a.replace('\t', '')
    a = a.replace(',', '')
    a = a.replace('.', '')
    a = a.replace(';', '')
    a = a.replace("'", '')
    a = a.replace('“', '')
    a = a.replace('”', '')
    a = a.replace('!', '')
    a = a.replace('’', ' ')
    a = a.replace('—', '')
    a = a.replace('-', '')
    a = a.replace('(', '')
    a = a.replace(')', '')
    a = a.strip()
    a = a.lower()
    if a:
        Novel.append(a)
```

```python
txt = ''
for i in Novel[27: ]:
    txt = txt+' ' + i
    
```

```python
words = [i for i in txt.split(' ') if i]
    
```

```python
len(words)
```

```python
wdict = {}

for i in words:
    if i in wdict:
        wdict[i] +=1
    else:
        wdict[i] = 1
```

```python
wdict
```

zipf

```python
wlist = sorted(list(wdict.items()), key=lambda x: x[1], reverse=True)
```

```python
wlist = [(i[0], i[1], k+1) for k, i in enumerate(wlist)]
```

```python
wlist[:10]
```

```python
import numpy as np

ws, freq, rank = np.array(wlist).T
freq = [int(i) for i in freq]
rank = [int(i) for i in rank]
```

```python
plt.style.use('fivethirtyeight')
plt.plot(rank,freq,'b-s')
plt.ylabel('$Frequency$', fontsize = 20)
plt.xlabel('$Rank$', fontsize = 20)
plt.xscale('log')
plt.yscale('log')
plt.title('Zipf Distribution', fontsize = 20)
plt.show()
```
